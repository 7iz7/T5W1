```python
import matplotlib.pyplot as plt

# Plotting histogram of TrafficIndexWeekAgo
plt.figure(figsize=(10, 6))
plt.hist(df['TrafficIndexWeekAgo'], bins=10, edgecolor='black', alpha=0.7)
plt.xlabel('Traffic Index Week Ago')
plt.ylabel('Frequency')
plt.title('Distribution of Traffic Index Week Ago')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Convert 'Datetime' column to datetime type (if not already done)
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Plotting the distribution of Datetime
plt.figure(figsize=(12, 6))
plt.hist(df['Datetime'], bins=50, edgecolor='black', alpha=0.7)
plt.xlabel('Datetime')
plt.ylabel('Frequency')
plt.title('Distribution of Datetime')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Convert 'Datetime' column to datetime type
df['Datetime'] = pd.to_datetime(df['Datetime'])

# Plotting Traffic Index over time
plt.figure(figsize=(12, 6))
plt.plot(df['Datetime'], df['TrafficIndexLive'], marker='o', linestyle='-', color='b', label='Traffic Index')
plt.xlabel('Datetime')
plt.ylabel('Traffic Index')
plt.title('Traffic Index Variation over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Plotting Jams Count and Jams Length over time
plt.figure(figsize=(12, 6))

plt.subplot(2, 1, 1)
plt.plot(df['Datetime'], df['JamsCount'], marker='o', linestyle='-', color='g', label='Jams Count')
plt.xlabel('Datetime')
plt.ylabel('Jams Count')
plt.title('Traffic Jams Count over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(df['Datetime'], df['JamsLength'], marker='o', linestyle='-', color='r', label='Jams Length')
plt.xlabel('Datetime')
plt.ylabel('Jams Length (km)')
plt.title('Traffic Jams Length over Time')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```

### Explore relationships between variables (e.g., Traffic_Index vs. Weather_Condition).


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming df is already defined as per previous example

# Generate hypothetical weather conditions
np.random.seed(0)
weather_conditions = ['Clear', 'Rainy', 'Cloudy', 'Snowy']
df['WeatherCondition'] = np.random.choice(weather_conditions, size=len(df))

# Explore relationship between TrafficIndexLive and WeatherCondition
plt.figure(figsize=(10, 6))

# Scatter plot with different colors for each weather condition
for weather in weather_conditions:
    subset = df[df['WeatherCondition'] == weather]
    plt.scatter(subset['TrafficIndexLive'], subset.index, label=weather, alpha=0.7)

plt.xlabel('Traffic Index')
plt.ylabel('Index')
plt.title('Traffic Index vs. Weather Condition')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping data by City and calculating mean JamsCount
city_jams_count = df.groupby('City')['JamsCount'].mean().sort_values(ascending=False)

# Plotting bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x=city_jams_count.index, y=city_jams_count.values, palette='viridis')
plt.xlabel('City')
plt.ylabel('Mean Jams Count')
plt.title('Mean Jams Count by City')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Counting occurrences of each city
city_counts = df['City'].value_counts()

# Plotting pie chart
plt.figure(figsize=(8, 8))
plt.pie(city_counts, labels=city_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Observations by City')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Violin plot for TrafficIndexWeekAgo by City
plt.figure(figsize=(12, 8))
sns.violinplot(x='City', y='TrafficIndexWeekAgo', data=df, palette='viridis')
plt.xlabel('City')
plt.ylabel('Traffic Index Week Ago')
plt.title('Distribution of Traffic Index Week Ago by City')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plotting histograms for TrafficIndexLive and JamsDelay
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['TrafficIndexLive'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Traffic Index Live')
plt.ylabel('Frequency')
plt.title('Distribution of Traffic Index Live')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(df['JamsDelay'], bins=20, edgecolor='black', alpha=0.7)
plt.xlabel('Jams Delay')
plt.ylabel('Frequency')
plt.title('Distribution of Jams Delay')
plt.grid(True)

plt.tight_layout()
plt.show()


```
